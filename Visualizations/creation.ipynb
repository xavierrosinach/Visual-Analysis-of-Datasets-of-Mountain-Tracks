```python
import pandas as pd
import folium
import os
import altair as alt

# Increase the altair maximum rows filter
alt.data_transformers.enable('default', max_rows=None)

data_path = '../../Data-2/canigo/Output-Data/Visualisations-Data-Frames'

dates_df = pd.read_csv(os.path.join(data_path, 'dates.csv'))
tracks_info = pd.read_csv(os.path.join(data_path, 'extended_tracks_info.csv'))

dates_df = dates_df[dates_df['year'] >= 2012]
tracks_info = tracks_info[tracks_info['year'] >= 2012]
```

```python
year_bar = alt.Chart(dates_df).mark_bar().encode(
    x=alt.X('sum(total_tracks):Q', title='Total Registered Tracks'),
    y=alt.Y('year:N', title='Year'),
    tooltip=[alt.Tooltip('year:N', title='Year'), alt.Tooltip('sum(total_tracks):Q', title='Total tracks')]
).properties(width=400, height=400)

month_bar = alt.Chart(dates_df).mark_bar().encode(
    x=alt.X('sum(total_tracks):Q', title='Total Registered Tracks'),
    y=alt.Y('month:N', title='Month', sort=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']),
    tooltip=[alt.Tooltip('month:N', title='Month'), alt.Tooltip('sum(total_tracks):Q', title='Total tracks')]
).properties(width=400, height=400)

year_bar | month_bar
```

```python
select_year = alt.selection_point(fields=['year'], on='click')

year_bar = alt.Chart(dates_df).mark_bar().encode(
    x=alt.X('sum(total_tracks):Q', title='Total Registered Tracks'),
    y=alt.Y('year:N', title='Year'),
    opacity=alt.condition(select_year, alt.value(1), alt.value(0.2)),
    tooltip=[alt.Tooltip('year:N', title='Year'), alt.Tooltip('sum(total_tracks):Q', title='Total tracks')]
).add_params(select_year
).properties(width=450, height=400)

month_bar = alt.Chart(dates_df).mark_bar().encode(
    x=alt.X('sum(total_tracks):Q', title='Total Registered Tracks'),
    y=alt.Y('month:N', title='Month', sort=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']),
    tooltip=[alt.Tooltip('month:N', title='Month'), alt.Tooltip('sum(total_tracks):Q', title='Total tracks')]
).add_params(select_year
).transform_filter(select_year
).properties(width=450, height=400)

evolution_line = alt.Chart(dates_df).mark_line().encode(
    x=alt.X('date:T', title='Date'),
    y=alt.Y('cumul_total_tracks:Q', title='Cumulative Sum of Registered Tracks')
).add_params(select_year
).transform_filter(select_year
).properties(width = 1000, title='Evolution of Registration of Tracks')

(year_bar | month_bar) & evolution_line
```
